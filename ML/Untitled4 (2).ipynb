```python
import pandas as pd

df = pd.read_csv("New_dataset.csv")
df.head()
```

```python
df.info()
df.describe()
df.columns
```

```python
df = df.drop_duplicates()
```

```python
df.info()
df.describe()
df.columns
```

```python
df = df[(df['Humidity'] >= 0) & (df['Humidity'] <= 100)]
df = df[df['Gas resistance in (Kohm)'] > 0]
df = df[(df['Red'] >= 0) & (df['Red'] <= 255)]
df = df[(df['Green'] >= 0) & (df['Green'] <= 255)]
df = df[(df['Blue'] >= 0) & (df['Blue'] <= 255)]
```

```python
df.info()
df.describe()
df.columns
```

```python
df['VOC_percent'] = (
    df['Difference']
    / df['BaseLine GR in(Kohm)']
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
df.head()
```

What this means: Mathematically, these two are currently providing identical information to your model. If $Difference = Baseline - Current$, then $VOC\% = Difference / Baseline$.Action: In your Random Forest model, you don't actually need both. Since VOC_percent is a ratio (normalized), it is generally superior because it scales better if your baseline shifts over different days or different sensors.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
```

```python
X = df[['Red', 'Green', 'Blue']]
```

```python
y_r = df['Ripeness']
y_c = df['Color']
y_ch = df['Chemical Used']
```

```python

X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X,
    y_r,
    test_size=0.2,
    random_state=42,
    stratify=y_r
)
```

```python
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X,
    y_c,
    test_size=0.2,
    random_state=42,
    stratify=y_c
)
```

```python
X_train_ch, X_test_ch, y_train_ch, y_test_ch = train_test_split(
    X,
    y_ch,
    test_size=0.2,
    random_state=42,
    stratify=y_ch
)
```

```python
l1_r = LogisticRegression(
    penalty='l1',
    solver='liblinear',
    max_iter=1000
)

l1_c = LogisticRegression(
    penalty='l1',
    solver='liblinear',
    max_iter=1000
)

l1_ch = LogisticRegression(
    penalty='l1',
    solver='liblinear',
    max_iter=1000
)
```

```python
l1_r.fit(X_train_r, y_train_r)

l1_c.fit(X_train_c, y_train_c)

l1_ch.fit(X_train_ch, y_train_ch)
```

```python
l2_r = LogisticRegression(
    penalty='l2',
    solver='liblinear',
    max_iter=1000
)

l2_c = LogisticRegression(
    penalty='l2',
    solver='liblinear',
    max_iter=1000
)

l2_ch = LogisticRegression(
    penalty='l2',
    solver='liblinear',
    max_iter=1000
)
```

```python
l2_r.fit(X_train_r, y_train_r)

l2_c.fit(X_train_c, y_train_c)

l2_ch.fit(X_train_ch, y_train_ch)
```

```python
pred_l1_r = l1_r.predict(X_test_r)
pred_l2_r = l2_r.predict(X_test_r)

pred_l1_c = l1_c.predict(X_test_c)
pred_l2_c = l2_c.predict(X_test_c)

pred_l1_ch = l1_ch.predict(X_test_ch)
pred_l2_ch = l2_ch.predict(X_test_ch)
```

```python
print("=== RIPENESS ===")
print("L1:", accuracy_score(y_test_r, pred_l1_r))
print("L2:", accuracy_score(y_test_r, pred_l2_r))

print("\n=== COLOR ===")
print("L1:", accuracy_score(y_test_c, pred_l1_c))
print("L2:", accuracy_score(y_test_c, pred_l2_c))

print("\n=== CHEMICAL ===")
print("L1:", accuracy_score(y_test_ch, pred_l1_ch))
print("L2:", accuracy_score(y_test_ch, pred_l2_ch))
```

```python
print(classification_report(
    y_test_r,
    pred_l1_r
))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test_ch, pred_l2_ch)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Chemical Confusion Matrix")

plt.show()
```
